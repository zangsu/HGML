# 캔 안에 들어있는 와인들을 분류하자.

## 데이터 준비

```python
import pandas as pd
wine = pd.read_csv('https://bit.ly/wine-date')
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.describe()
```

```python
data = wine.drop('class', axis = 1).to_numpy()
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target, random_state = 42, test_size = 0.2) 
```

```python
print(train_input.shape, test_input.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
print(lr.coef_, lr.intercept_)
```

# 결정트리를 사용해보자.

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 42)
dt.fit(train_scaled, train_target)
print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10, 7))
plot_tree(dt)
plt.show()
```

```python
#엄청난 트리를 좀 더 자세히 보기위해 깊이를 제한
plt.figure(figsize=(20, 14))
plot_tree(dt, max_depth = 2, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

#결정트리의 설명

결정트리는 각각의 조건에 따라 분기를 만들면서 리프노드로 뻗어간다.

각각의 인풋은 조건을따라 특정 리프노드로 도착하는데, 그곳의 노드에서 존재하는 더 많은 클래스로 분류가 된다.

+ gini : 불순도 (각 노드에서 클래스의 비율차이가 얼마나 많이 나는지)

결정트리는  자식으로 갈수록 불순도가 높아지는 방향으로 가지를 뻗어간다

```python
#가지치기
dt = DecisionTreeClassifier(max_depth = 3, random_state = 42)
dt.fit(train_scaled, train_target)
print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
plt.figure(figsize=(20, 15))
plot_tree(dt, filled=True, feature_names = ['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
#위 트리를 가지고 분석을 할 수 있다.
#사실 결정트리를 만들때는 데이터 스케일링이 필요없다.
dt = DecisionTreeClassifier(max_depth = 3, random_state = 42)
dt.fit(train_input, train_target)
print(dt.score(train_input, train_target))
print(dt.score(test_input, test_target))
```

```python
plt.figure(figsize=(20, 15))
plot_tree(dt, filled=True, feature_names = ['alcohol', 'sugar', 'pH'] )
plt.show()
```

```python
#결정트리를 사용해서 각 특성의 중요도를 확인할 수 있다.
print(dt.feature_importances_)
```
